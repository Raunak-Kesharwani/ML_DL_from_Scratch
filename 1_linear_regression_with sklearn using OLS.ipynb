```python
import numpy as np 
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    "x" :[1,2,3],
    "y" : [1,3,4]
})
```

```python
X = df["x"].values.reshape(-1,1)
y = df["y"].values
```

```python
lr = LinearRegression()
```

```python
X.reshape(-1,1)
```

```python
lr.fit(X,y)
```

```python
y_pred = lr.predict(X)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(df["x"] , df["y"])


plt.plot(X, y_pred , color = "red")
plt.xlabel("X")
plt.ylabel("y")
```

```python
# R2_score
lr.score(X , y)
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(df["cgpa"] , df["package"] , test_size=0.2 , random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train.values.reshape(-1,1) , y_train)
```

```python
# R2_Score for training 
lr.score(X_train.values.reshape(-1,1) , y_train)
```

```python
# R2_score for testing 
lr.score(X_test.values.reshape(-1,1) , y_test)
```

```python
y_train_pred = lr.predict(X_train.values.reshape(-1,1))
y_test_pred = lr.predict(X_test.values.reshape(-1,1))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(12, 5))   # 1 row, 2 columns

# ---------------- LEFT PLOT (TRAIN) ----------------
ax[0].scatter(X_train, y_train, label="Train Data")
ax[0].plot(X_train, y_train_pred, color="green", label="Train Prediction")
ax[0].set_title("Training Data vs Prediction")
ax[0].set_xlabel("X_train")
ax[0].set_ylabel("y_train")
ax[0].legend()
ax[0].grid(True)

# ---------------- RIGHT PLOT (TEST) ----------------
ax[1].scatter(X_test, y_test, label="Test Data")
ax[1].plot(X_test, y_test_pred, color="orange", label="Test Prediction")
ax[1].set_title("Test Data vs Prediction")
ax[1].set_xlabel("X_test")
ax[1].set_ylabel("y_test")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
